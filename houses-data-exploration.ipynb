```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./datasets/melb_data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.dropna(axis=0)
```

```python
df.describe()
```

```python
prediction_target = df['Price']
prediction_target.head()
```

```python
for column in df.columns:
    dtype = df[column].dtype

    if dtype=='float64' or dtype=='int64':
        correlation = df['Price'].corr(df[column])
        print(f"Correlation of {column} to Price is {correlation}")
```

```python
prediction_features = ['Rooms', 'Bathroom', "BuildingArea"]
```

```python
df_prediction = df[prediction_features]
```

```python
df_prediction.describe()
```

```python
df_prediction.head()
```

```python
Model = DecisionTreeRegressor(random_state=1)
Model.fit(df_prediction, prediction_target)
```

```python
df_prediction.head()
```

```python
predicted_prices = Model.predict(df_prediction).round()
predicted_prices
```

```python
df_prediction['Predicted price'] = predicted_prices
```

```python
df_prediction.head()
```
