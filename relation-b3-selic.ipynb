```python
import pandas as pd
df_indicators = pd.read_csv('./datasets/economic_indicators.csv')
df_bovespa = pd.read_csv('./datasets/bovespa_stocks.csv')
```

```python
df_bovespa.head()
```

```python
df_indicators.head()
```

```python
df_indicators['Date'] = pd.to_datetime(df_indicators['Date']).dt.strftime('%Y-%m-%d')

```

```python
df_bovespa['Date'] = pd.to_datetime(df_bovespa['Date'], errors='coerce').dt.date
```

```python
df_bovespa['Taxa Selic'] = ''
```

```python
for index, row in df_bovespa.loc[df_bovespa['Taxa Selic']==''].iterrows():
    condition = df_indicators['Date'] == row['Date']
    try:
        daily_indicator = df_indicators.loc[condition].head(1).to_dict(orient='records')[0]
        df_bovespa.at[index, 'Taxa Selic'] = daily_indicator['Taxa Selic']
    except:
        print(f"Error at date {row['Date']}")
```

```python
df_bovespa.loc[df_bovespa['Taxa Selic'].str.strip().ne('')]
```

```python
new_df_bovespa = df_bovespa.loc[df_bovespa['Taxa Selic'].str.strip().ne('')]
```

```python
new_df_bovespa.to_csv('./datasets/new_bovespa_stocks.csv')
```
